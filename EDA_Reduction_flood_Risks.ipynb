Cocatonate three datasets togother

```python
import glob
```

```python
df_all= glob.glob("*.csv")
```

```python
print(df_all)
```

```python
df_all = df1.merge(df2,on='date', how='left')
df_all = df_all.merge(df3,on='date', how='left')
```

```python
df_all.head()
```

```python
df_all.shape
```

```python
df_all.describe(include='all').style.background_gradient(axis=None, cmap='YlOrRd')
```

Cleaning data and visualize missing values

```python
df_all.isnull().sum()
```

```python
msno.bar(df_all,figsize=(10,7),fontsize=15)
plt.title('percentage of missing values per columns')
plt.show()
```

```python
df_all.columns
```

```python
#filling miss values 
#filling in waterlevel_y
df_all.fillna(df_all['waterlevel_y'].mean(), inplace=True)
#filling in waterlevel
df_all.fillna(df_all['waterlevel'].mean(), inplace=True)
```
